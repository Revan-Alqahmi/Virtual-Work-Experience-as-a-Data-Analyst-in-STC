```python
"""
TODO build your Recommender system to Highlight Programs that usesrs might be interested in
"""

# Importing necessary libraries
import pandas as pd
from surprise import Dataset, Reader, KNNBasic
from surprise.model_selection import train_test_split
from collections import defaultdict

# Load the dataset
data = pd.read_csv("/content/stc TV Data Set_T3.csv")

# Create a Surprise Reader object
reader = Reader(rating_scale=(1, 5))

# Load data into Surprise Dataset
surprise_data = Dataset.load_from_df(data[['user_id_maped', 'program_name', 'rating']], reader)

# Split the data into train and test sets
trainset, testset = train_test_split(surprise_data, test_size=0.2)

# Build the collaborative filtering model
model = KNNBasic(sim_options={'user_based': False})  # User-based or item-based, depending on your preference

# Train the model
model.fit(trainset)

# Get recommendations for user 26138
user_id = 26138
program_ids = data['program_name'].unique()

# A dictionary to store recommendations
recommendations = defaultdict(list)

for program_id in program_ids:
    prediction = model.predict(user_id, program_id)
    recommendations[program_id] = prediction.est

# Sort recommendations by predicted rating
sorted_recommendations = sorted(recommendations.items(), key=lambda x: x[1], reverse=True)

# Print top 5 recommended programs
top_5_recommendations = sorted_recommendations[:5]
for program, rating in top_5_recommendations:
    print(f"Program: {program}, Predicted Rating: {rating}")
```

```python
"""
TODO show the recommendations (top 5) for the people who watched "Moana" movie
"""

# Importing necessary libraries
import pandas as pd
from surprise import Dataset, Reader, KNNBasic
from surprise.model_selection import train_test_split
from collections import defaultdict

# Load the dataset
df = pd.read_csv("/content/stc TV Data Set_T3.csv")

# Filter the dataset for users who watched "Moana"
moana_users = df[df['program_name'] == 'Moana']['user_id_maped'].unique()

# Load the data into Surprise format
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(df[['user_id_maped', 'program_name', 'rating']], reader)

# Split the data into train and test sets
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

# Train the KNNBasic model
sim_options = {
    'name': 'msd',
    'user_based': True  # Use user-based collaborative filtering
}
algo = KNNBasic(sim_options=sim_options)
algo.fit(trainset)

# Dictionary to store top 5 recommendations for each user
top_n = defaultdict(list)

# Get recommendations for each user who watched "Moana"
for user in moana_users:
    # Check if the user exists in the training set
    if user in trainset._raw2inner_id_users:
        # Get top 5 recommendations for each user
        try:
            recommendations = algo.get_neighbors(trainset.to_inner_uid(user), k=5)
            top_n[user] = recommendations
        except IndexError as e:
            print(f"Error: {e}")
    else:
        print(f"User {user} not found in the training set.")

# Print the top 5 recommendations for each user
for user, recommendations in top_n.items():
    print(f"Top 5 recommendations for User {user}: {recommendations}")
```
